```python
fingerprint_origin["SEGMENT_ID"] = fsc_cash_flow_fact.groupby("ACCOUNT_KEY").sum()['CURRENCY_AMOUNT']
```

```python
fingerprint_origin.columns = ["TotalNoOfTran", "TotalAmountOfTran"]
fingerprint_origin.head()
```

```python
max_tran = fsc_cash_flow_fact.groupby("ACCOUNT_KEY").max()["CURRENCY_AMOUNT"]
min_tran = fsc_cash_flow_fact.groupby("ACCOUNT_KEY").min()["CURRENCY_AMOUNT"]
mean_tran = fsc_cash_flow_fact.groupby("ACCOUNT_KEY").mean()["CURRENCY_AMOUNT"]
#std_tran = fsc_cash_flow_fact.groupby("ACCOUNT_KEY").std()["CURRENCY_AMOUNT"]
```

```python
fingerprint_origin["MaxTran"] = max_tran
fingerprint_origin["MinTran"] = min_tran
fingerprint_origin["MeanTran"] = mean_tran
#fingerprint_origin["StdTran"] = std_tran
```

```python
fingerprint_origin.head()
```

```python
def convert_categorical_by_acccount(category):
    dummy_by_account = fsc_cash_flow_fact.groupby(["ACCOUNT_KEY", category]).\
                            count().reset_index().\
                            pivot_table(values=["TRANSACTION_KEY"], #最初のカラムを選択している
                                        index=["ACCOUNT_KEY"],columns=[category],
                                        aggfunc="sum", fill_value=0)
    return dummy_by_account
```

```python
tran_type = convert_categorical_by_acccount("TRANSACTION_TYPE_KEY")
```

```python
#tran_type = fsc_cash_flow_fact.groupby(["ACCOUNT_KEY", "TRANSACTION_TYPE_KEY"]).\
#                            count().reset_index().\
#                            pivot_table(values=["TRANSACTION_KEY"], 
#                                        index=["ACCOUNT_KEY"],columns=['TRANSACTION_TYPE_KEY'],
#                                        aggfunc="sum", fill_value=0)
```

```python
branch_type = convert_categorical_by_acccount('BRANCH_KEY')
branch_type.head()
```

```python
fingerprint_origin.drop(accout_drop_list, axis=1,inplace=True)
```

```python
need_dummy_for_account_cols = [s for s in account_cols if s not in accout_drop_list]
```

```python
for col in need_dummy_for_account_cols:
    dummy = pd.get_dummies(fingerprint_origin[col])
    fingerprint_origin.drop(col, axis=1, inplace=True)
    fingerprint_origin = pd.concat([fingerprint_origin, dummy], axis=1)
```

## PCA＋クラスタリング

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
fingerprint_ss = ss.fit_transform(fingerprint_origin)
```

```python
pca = PCA(n_components=0.95)
```

```python
fingerprint_pca = pca.fit_transform(fingerprint_ss)
```

```python
fingerprint_pca.shape
```

```python
plt.scatter(fingerprint_pca[:,0], fingerprint_pca[:, 1])
```

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2, random_state=0)
```

```python
fingerprint_tsne = tsne.fit_transform(fingerprint_ss)
```

```python
plt.scatter(fingerprint_tsne[:,0], fingerprint_tsne[:, 1])
```

```python
#上記で何かしらの塊がある
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=2, random_state=1)
y_cluster = km.fit_predict(fingerprint_tsne)
```

```python
set(y_cluster)
```

```python
plt.scatter(fingerprint_tsne[:,0], fingerprint_tsne[:, 1], c=y_cluster)
plt.legend()
```
